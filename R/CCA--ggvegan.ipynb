```r
library("vegan")
library("ggvegan")
library("cowplot")
library("ggrepel")
data(varespec, varechem)
vare.cca <- cca(varespec, varechem)
```

```r
obj <- fortify(vare.cca)
want <- obj$Score %in% c("species", "sites")

autoplot(vare.cca) +  geom_text(data = obj[want, , drop = FALSE ],
                      aes(x = Dim1, y = Dim2, label = Label,
                          colour = Score))
```

```r
autoplot(vare.cca) +  geom_text_repel(data = obj[want, , drop = FALSE ],
                      aes(x = Dim1, y = Dim2, label = Label,
                          colour = Score))
```
